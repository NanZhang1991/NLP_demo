# word to xml

```python
import zipfile
templateDocx = zipfile.ZipFile("data/word/Template.docx")
newDocx = zipfile.ZipFile("data/word/NewDocument.docx", "a")
with open(templateDocx.extract("word/document.xml", "data/word"), encoding='utf-8') as tempXmlFile:
    tempXmlStr = tempXmlFile.read(
    )
replaceText={'关于成立':'这是替换的文本'}  
for key in replaceText.keys():
    tempXmlStr = tempXmlStr.replace(str(key), str(replaceText.get(key)))

with open("data/word/word/temp.xml", "w+", encoding='utf-8') as tempXmlFile:
    tempXmlFile.write(tempXmlStr)
    
for file in templateDocx.filelist:
    if not file.filename == "word/document.xml":
        newDocx.writestr(file.filename, templateDocx.read(file))

newDocx.write("data/word/temp.xml", "word/document.xml")

templateDocx.close()
newDocx.close()
```
